```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data

csv_file_path = 'data/scraped_ekkofilm_reviews.csv'
df = pd.read_csv(csv_file_path)
```

```python
ratings = df['Rating']

mean = ratings.mean()
median = ratings.median()
std_dev = ratings.std()

plt.figure(figsize=(10, 6))

# KDE plot to resemble a bell curve
sns.kdeplot(ratings, fill = True, color = "skyblue", linewidth = 2)

# Add lines for mean, median, and standard deviations
plt.axvline(mean, color = 'blue', linestyle = '--', linewidth = 1.5, label = f'Mean: {round(mean, 2)}')
plt.axvline(median, color = 'green', linestyle = '-', linewidth = 1.5, label = f'Median: {round(median, 2)}')
plt.axvline(mean - std_dev, color = 'orange', linestyle = '--', linewidth = 1.5, label = f'-1 SD: {round(mean - std_dev, 2)}')
plt.axvline(mean + std_dev, color = 'orange', linestyle = '--', linewidth = 1.5, label = f'+1 SD: {round(mean + std_dev, 2)}')

plt.title('Rating distribution', fontsize = 16)
plt.xlabel('Rating', fontsize = 14)
plt.ylabel('Density', fontsize = 14)
plt.legend()
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.tight_layout()

plt.show()
```
